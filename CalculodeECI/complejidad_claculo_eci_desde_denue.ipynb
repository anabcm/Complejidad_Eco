```python
import pandas as pd
import sys
import numpy as np
import matplotlib.pyplot as plt

from rca_matrix import *
from complexity import *
```

```python
denue = pd.read_csv('conteos_denue_1120.csv') # Cuidado, aquí debe ir el nombre de la base como esté guardada en sus ambientes

denue['cve_ent'] = denue.cve_ent.apply(lambda l: str(l).zfill(2))
denue['cve_mun'] = denue.cve_mun.apply(lambda l: str(l).zfill(3))
denue['cvegeo'] = denue['cve_ent']+denue['cve_mun']

denue = denue.rename(columns={'codigo_act':'clave_actividad','conteo':'ue'})
denue = denue.groupby(['cvegeo','clave_actividad'], as_index=False).sum()

print(denue.shape)
display(denue.head(10))
```

```python
matriz_denue = denue.pivot(index='cvegeo',columns='clave_actividad', values='ue')

print(matriz_denue.shape)
display(matriz_denue.head(10))
```

```python
matriz_denue = rca(matriz_denue)
eci, aci = complexity(matriz_denue)
```
